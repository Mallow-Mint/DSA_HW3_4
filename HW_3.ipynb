## Ex. 4 - Knight's Tour

The set of legal knight moves per square on a chess board defines a graph known as the knight's graph.  
A tour is a sequence of edges/vertices that, if traversed, visits each vertex exactly once (i.e. without doubling back).  


```python
### DO: Make a function to find a sequence of knight moves on an $8\times 8$ chess board s.t. all squares are visited only once.  
### Let it start from any square on the board.  
### The tour may be open or closed (i.e. it does not need to loop back to itself as in TSP). 

def initboard(size):
    board = np.full((size, size), 0, dtype=int)
    return board

def knights_tour(y, x, board, moves):
    size = len(board)
    knight = [(1, 2), (2, 1), (-1, 2), (-2, 1), (1, -2), (2, -1), (-2, -1), (-1, -2)]
    def valid_check(y, x, board):
        if 0 <= x < size and 0 <= y < size and board[y][x] == 0:
            return True
        else:
            return False
    def count_moves(y, x, board):
        count = 0
        for i, j in knight:
            ny, nx = y+i, x+j
            if valid_check(ny, nx, board):
                count += 1
        return count
    board[y][x] = moves
    if moves == size ** 2:
        return True
    next_move = []
    for i, j in knight:
        ny, nx = y + i, x + j
        if valid_check(ny, nx, board):
            move_forward = count_moves(ny, nx, board)
            next_move.append(((ny, nx), move_forward))
    next_move.sort(key=lambda x:x[1])
    for (ny,nx), _ in next_move:
        if knights_tour(ny, nx, board, moves+1):
            return True
        board[ny][nx]=0
    return False
```

```python
# DO: Prove that your method visits all squares exactly once

for i in range(20):
    board = initboard(8)
    start_x, start_y = np.random.randint(0, 8), np.random.randint(0, 8)
    print("Starting x and y:", start_x+1, "&", start_y+1) # so coordinates go from 1-8 instead of 0-7
    if knights_tour(start_y, start_x, board, 1):
        print("Knight's Tour found:")
        print(board)
        print("Is there exactly one instance of each number 1-64?", np.array_equal(np.sort(board.flatten()), np.arange(1, 65)))
        print("8*8=64, is there any number higher than 64 present?", np.any(board>64))
    else:
        print("No tour found.")
```
